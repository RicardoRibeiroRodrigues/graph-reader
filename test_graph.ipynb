```python
import matplotlib.pyplot as plt
import csv
```

```python
# Read 'graph_data.csv'
with open('data.csv', 'r') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    x = []
    y = []
    for row in csv_reader:
        x.append(float(row[0]))
        y.append(float(row[1]))

plt.figure(figsize=(10, 6))
# Scatter plot
plt.plot(x, y)
# plt.scatter(x, y)
plt.grid(True)
plt.show()
```
